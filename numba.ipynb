# NUMBA:
Numba is just in time compiler for python that works on best on code that uses
Numpy arrays and functions , and loops.

Because we all know that python is interpreted language and other programming languages like; C, C++ , JAVA are much faster than python since they are just in compiler languages, that means they convert the compiled code into machine level code and that code runs on machine level code , it leads to fast execution.

so that's why if you want your python code to be fast then you have to use NUMBA library which will give you the leverage like other programming languages.

```python
from numba import jit
import numpy as np
import time


x=np.arange(1000).reshape(100,10)

@jit(nopython=True)
def go_fast(a): #Function is compiled and runs in machine code
    trace=0.0
    for i in range(a.shape[0]):
        trace += np.tanh(a[i,i])#Try to find out tangent
    return a + trace

#Do Not report this.... Complition time is included in the execution time!
start = time.time()
go_fast(x)
end=time.time()
print('Elapsed (with complition) = %s ' %(end - start))

#Now function is compiled , re-time it executing  from cache
start=time.time()
go_fast(x)
end=time.time()
print('Elapsed (after complition) = %s ' %(end - start))
```
